```python
feature_train_before = {}
feature_vina_score = {}

with open('/mnt/processed_data/feature_test.txt','r') as fi:
    feature_trainlist =  [l[:-1] for l in fi if l.strip()]
    for rfpair_ in feature_trainlist:
        rfpair = rfpair_.split(',')
        feature_train_before[rfpair[0]]=rfpair_
with open('/mnt/processed_data/vina_feature_test.txt','r') as fi:
    vina_score_list =  [l[:-1] for l in fi if l.strip()]
    for groundtruthpair in vina_score_list:
        groundtruthpair = groundtruthpair.split(' ')
        feature_vina_score[str(groundtruthpair[0])] = groundtruthpair[1]
```

```python
total_feature = {}
for prot in feature_train_before.keys():
    if prot not in feature_vina_score.keys():
        continue
    total_feature[prot] = feature_train_before[prot]+','+feature_vina_score[prot]
output_file = '/mnt/processed_data/full_feature_test.txt'
with open(output_file,'w') as fo:
    for protein in total_feature.keys():
        fo.write(total_feature[protein]+'\n')
```
